```python
# Caluclo el rango intercuartil para ver los datos atipicos

Q1 =df['Total Interactions'].quantile(0.25)
Q3 =df['Total Interactions'].quantile(0.75)

IQR = Q3-Q1

lower_bound = Q1 -1.5*IQR
upper_bound =Q3 +1.5*IQR

outliers = df[(df['Total Interactions'] < lower_bound) | (df['Total Interactions'] > upper_bound)]


outliers
```

```python
num_outliers = len(outliers)

num_outliers
```
